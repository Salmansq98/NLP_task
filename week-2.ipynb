```python
# importing some vital libaries.
import tensorflow as tf

import tensorflow_datasets as tfds
```

```python
# loading the dataset from tensorflow_dataset.
imdb, info = tfds.load('imdb_reviews', with_info=True, as_supervised=True)
```

```python
import numpy as np

# spliting the dataset.
train_data, test_data = imdb['train'], imdb['test']

training_sentences = []
training_labels=[]

testing_sentences=[]
testing_labels=[]

# adding the sentences and labels to specific list.
for s,l in train_data:
    training_sentences.append(s.numpy().decode('utf-8'))
    training_labels.append(l.numpy())
    
for s,l in test_data:
    testing_sentences.append(s.numpy().decode('utf-8'))
    testing_labels.append(l.numpy())

# converting the list into array.
training_labels_final = np.array(training_labels)
testing_labels_final = np.array(testing_labels)
```

```python
vocab_size = 10000 
embedding_dim = 16
max_length = 120
trunc_type = 'post'
oov_tok = '<OOV>'
# importing the tokenizer and pad_sequences module.
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

tokenizer = Tokenizer(num_words=vocab_size, oov_token=oov_tok)
```

It will take top-n no's of words by volume.
oov_token = it will take the distinct(which are not in the word_index) words from the corpus and it will assign oov_tok to it. num_words = 10000, but it will take only 9999 words, 'cause 1 word is reserved for oov_token.

```python
# its gonna built 'dict', which will consist of word_index and a owrd.
# It will map a word to a number.
tokenizer.fit_on_texts(training_sentences)

# return the word_index.
word_index = tokenizer.word_index

# it will convert the text into sequence.
sequences = tokenizer.texts_to_sequences(training_sentences)

## Notice in the sentence, we have sentences which are not equal. I'll
# use pad_sequence to make it equal size. 
# truncate the exceeded sentences from the end, which are above maxlen 
padded = pad_sequences(sequences, maxlen=max_length, truncating=trunc_type)

# test data.
testing_sequence = tokenizer.texts_to_sequences(testing_sentences)
test_padding = pad_sequences(testing_sequence, maxlen=max_length, truncating=trunc_type)
```

### Creating the model.

```python
from tensorflow.keras.layers import Embedding, Flatten, Dense

model = tf.keras.models.Sequential([
    Embedding(vocab_size, embedding_dim, input_length=max_length),
    Flatten(), # it will convert multi-dimension i/p into single dimension
    Dense(6, activation='relu'),  # dense unit with 6 neurons. 
    Dense(1, activation='sigmoid') # dense unit with 1 neurons.
])

# compile the model with loss(binary_crossentropy), optimizer(adam) and metrics(accuracy)
model.compile(loss = 'binary_crossentropy', optimizer = 'adam',
             metrics=['accuracy'])

model.summary() # returns the model summary
```

```python
# getting the first layer.
e = model.layers[0] 

# getting the weights of that layer.
weights = e.get_weights()[0]
# printing the shape.
print(weights.shape)
```

```python
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])
```

```python
 #importing the input and output lib.
import io 

# opening a file vecs.tsv and in the write mode.
out_v = io.open('vecs.tsv','w', encoding='utf-8')
# opening a file meta.tsv and in the write mode.
out_m = io.open('meta.tsv', 'w', encoding='utf-8')

for word_num in range(1, vocab_size):
    word = reverse_word_index[word_num] # take a word  
    embeddings = weights[word_num] # take the index of that word
    out_m.write(word+'\n') # write the assigned word in the file.
    
    out_v.write('\t'.join([str(x) for x in embeddings]) + "\n")

# closing the files and saving it.
out_v.close() 
out_m.close()
```
